```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('twitchdata-update.csv')

df.head()
```

```python
print(df.isnull().sum())
lang_stats = df.groupby('Language')['Followers gained'].mean().sort_values(ascending = False)
print(lang_stats)
```

```python
ax=lang_stats.head(10).plot(kind='bar',color='purple')
plt.title('Top 10 Languages by Average Followers Gained')
plt.ylabel('Followers')
for container in ax.containers:
    ax.bar_label(container, fmt=lambda x: f'{x/1000:.1f}K')

plt.show()
```

```python
maturity_stats = df.groupby('Mature')['Followers gained'].mean()
print(maturity_stats)
```
